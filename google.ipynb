```python
graph = {}
    
#undirected_graph
def add_edge(graph, parent, child):
    if child in graph and parent in graph:
        graph[parent].append(child)
        graph[child].append(parent)
    elif child in graph:
        graph[child].append(parent)
        graph[parent] = [child]
    elif parent in graph:
        graph[parent].append(child)
        graph[child] = [parent]
    else:
        graph[child] = [parent]
        graph[parent] = [child]
        

#directed_graph
def add_edge_directed(graph, parent, child):
    if child in graph and parent in graph:
        graph[parent].append(child)
    elif child in graph:
        graph[parent] = [child]
    elif parent in graph:
        graph[parent].append(child)
        graph[child] = []
    else:
        graph[child] = []
        graph[parent] = [child]

add_edge(graph, 0, 1)
add_edge(graph, 0, 2)
add_edge(graph, 0, 3)
add_edge(graph, 2, 5)
add_edge(graph, 3, 4)
add_edge(graph, 4, 6)
add_edge(graph, 6, 7)
add_edge(graph, 1, 7)

#returning the shortest path using dfs
def shortest_path_using_minvar(graph, start, end, path_final, path = []):
    path = path + [start]
    
    if start == end:
        if path_final[0] == None or len(path_final) > len(path):
            path_final.append(path)
            path_final.pop(0)
        return

    for node in graph[start]:
        if node not in path:
            shortest_path_using_minvar(graph, node, end, path_final, path)
            
            
path_final = [None]
shortest_path_using_minvar(graph, 0, 7, path_final)
print(path_final)

```
